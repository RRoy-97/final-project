## Extract CSVs into DataFrames

```python
# Import csv into loan_df
loan_file = "Resources/loan.csv"
loan_df = pd.read_csv(loan_file )
#loan_df
```

```python
loan_transformed_df = loan_df.drop(["collection_recovery_fee", 
                                 "inq_last_6mths",
                                 "term_mnths",
                                 "mths_since_last_delinq", 
                                 "mths_since_last_major_derog", 
                                 "policy_code",
                                 "funded_amnt_inv",
                                 "pub_rec",
                                 "installment",
                                 "tot_coll_amt",
                                 "verification_status_joint",
                                 "delinq_2yrs",
                                 "annual_inc_joint",
                                 "mths_since_last_record",
                                 "initial_list_status",
                                 "out_prncp_inv",
                                 "out_prncp",
                                 "total_pymnt_inv",
                                 "last_credit_pull_d",
                                 "next_pymnt_d",
                                 "collections_12_mths_ex_med",
                                 "emp_title",
                                 "title",
                                 "mths_since_rcnt_il",
                                 "total_bal_il",
                                 "il_util",
                                 "max_bal_bc",
                                 "all_util",
                                 "total_rev_hi_lim",
                                 "inq_fi",
                                 "total_cu_tl",
                                 "inq_last_12m",
                                 "total_rec_late_fee",
                                 "url",
                                 "earliest_cr_line",
                                 "open_acc",
                                 "revol_util",
                                 "revol_bal",
                                 "total_acc",
                                 "total_pymnt",
                                 "total_rec_int",
                                 "total_rec_prncp",
                                 "verification_status",
                                 "pymnt_plan",
                                 "open_acc_6m",
                                 "open_il_6m",
                                 "open_il_12m",
                                 "open_il_24m",
                                 "open_rv_12m",
                                 "open_rv_24m",
                                 "zip_code",
                                 "dti_joint",
                                 "acc_now_delinq",
                                 "acc_now_delinq"], 
                                axis=1).reset_index()
loan_transformed_df = loan_transformed_df.drop(["index"], axis=1)
loan_transformed_df
```

# Clean dataset

```python
# replacing na values in desc with none
loan_transformed_df["desc"].fillna("none", inplace = True)
```

```python
# Drop rows where emp_length has null value
loan_transformed_df.dropna(subset=['emp_length'], how='all', inplace=True)
```

```python
# Find missing values for each column
loan_transformed_df.isnull().sum()
```

```python
#List unique values in the df['loan_status'] column
loan_transformed_df.loan_status.unique()
```

```python
# Count number of each loan status
loan_status = loan_transformed_df['loan_status'].value_counts()
loan_status
```

```python
# Drop unwanted loan status'
a = ['Late (31-120 days)', 'In Grace Period', 'Late (16-30 days)',
       'Does not meet the credit policy. Status:Fully Paid',
       'Does not meet the credit policy. Status:Charged Off', 'Issued']

loan_transformed_df = loan_transformed_df[~loan_transformed_df['loan_status'].isin(a)]

loan_transformed_df
```

```python
loan_transformed_df['loan_status'].value_counts()
```

```python
# Isolate and remove application type = 'JOINT'
jointAppl = loan_transformed_df[loan_transformed_df['application_type'] == 'JOINT'].index

loan_transformed_df.drop(jointAppl , inplace=True)
```

```python
loan_transformed_df.dtypes
```

```python
#Export cleaned data to csv
loan_transformed_df.to_csv('Resources/loan_transformed.csv')
```
